```python
import re
import urllib.request
import json
import ast
import pandas as pd
import numpy as np

URL = "https://covid19.saglik.gov.tr/TR-66935/genel-koronavirus-tablosu.html"

contents = urllib.request.urlopen(URL).read().decode('utf-8')
x = re.search("var geneldurumjson = \[(.*?)];//", contents).group()

x = x[21:]
x = x[:-3]


all_list = ast.literal_eval(x)

confirmed = []
death = []
recovered = []
daily_confirmed = []
daily_death = []
daily_recovered = []
date = []
test = []



for i in all_list[::-1]:
    date.append(i['tarih'])
    daily_recovered.append(i['gunluk_iyilesen'].replace('.', ''))
    daily_death.append(i['gunluk_vefat'].replace('.', ''))
    daily_confirmed.append(i['gunluk_vaka'].replace('.', ''))
    recovered.append(i['toplam_iyilesen'].replace('.', ''))
    death.append(i['toplam_vefat'].replace('.', ''))
    confirmed.append(i['toplam_hasta'].replace('.', ''))
    test.append(i['gunluk_test'].replace('.', ''))
print("done")

daily_recovered = ['0' if i == '' else i for i in daily_recovered]
daily_death = ['0' if i == '' else i for i in daily_death]
daily_confirmed = ['0' if i == '' else i for i in daily_confirmed]
recovered = ['0' if i == '' else i for i in recovered]
death = ['0' if i == '' else i for i in death]
confirmed = ['0' if i == '' else i for i in confirmed]
test = ['0' if i == '' else i for i in test]


date = np.array(date)
daily_recovered = np.array(daily_recovered, dtype=np.int32)
daily_death = np.array(daily_death, dtype=np.int32)
daily_confirmed = np.array(daily_confirmed, dtype=np.int32)
recovered = np.array(recovered, dtype=np.int32)
death = np.array(death, dtype=np.int32)
confirmed = np.array(confirmed, dtype=np.int32)
test = np.array(test, dtype=np.int32)



df = pd.DataFrame()


df['Date'] = date
df['Confirmed'] = confirmed
df['Recovered'] = recovered
df['Death'] = death
df['Daily_Confirmed'] = daily_confirmed
df['Daily_Recovered'] = daily_recovered
df['Daily_Death'] = daily_death
df['Test'] = test

df['Date'] = pd.to_datetime(df['Date'])
df
```

```python
df['Daily_Confirmed_Format'] = df['Confirmed'] - df['Confirmed'].shift(1)
df.loc[274, 'Daily_Confirmed_Format'] = df.loc[274, 'Daily_Confirmed']
```

```python
import matplotlib.pyplot as plt
plt.plot(df['Daily_Confirmed_Format'])
```

```python
df['New_Deaths'] = df['Death'] - df['Death'].shift(1)
```

```python
df['New_Deaths'] = df['New_Deaths'].fillna(0)
```

```python
df['New_Recovered'] = df['Recovered'] - df['Recovered'].shift(1)
df['New_Recovered'] = df['New_Recovered'].fillna(0)
```

```python
df['Daily_Confirmed_Format'] = df['Daily_Confirmed_Format'].fillna(1)
```

```python
df.loc[276, 'New_Recovered'] = df.loc[276, 'Daily_Recovered']
```

```python
df = df[['Date', 'Confirmed', 'Recovered', 'Death', 'Test', 'Daily_Confirmed_Format', 'New_Deaths', 'New_Recovered']]
df
```

```python
df.rename(columns={'Daily_Confirmed_Format': 'Daily_Confirmed', 'New_Deaths': 'Daily_Death', 'New_Recovered': 'Daily_Recovered'}, inplace=True)
```
